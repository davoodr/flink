```python
%matplotlib inline

import pandas
import matplotlib.pyplot as plt

csv = pandas.read_csv('benchmarkX/total.csv', sep=',', names=["delay", "index", "name", "cumCount"])
sort = csv.sort_values(['index'])

fig = plt.figure(figsize=(15,10))
fig.suptitle('Latency for each element without migration', fontsize=18, fontweight='bold')
fig.subplots_adjust(top=0.85)

ax = fig.add_subplot(111)
ax.set_title('Latency for each element without migration', fontsize=18, fontweight='bold')
ax.set_xlabel('Element index')
ax.set_ylabel('Latency in milliseconds')

# TODO Duplicated indices!
plt.plot(sort['index'], sort['delay'], 'r-')
plt.show()

fig.set_size_inches(18.5, 10.5)
fig.savefig('test2png.png', dpi=100)
```

```python
%matplotlib inline

import pandas
import matplotlib.pyplot as plt

csv = pandas.read_csv('blub/throughput.txt', sep=',', names=["instance", "throughput", "timestamp"])
csv = csv.sort_values(['timestamp'])

instance1 = csv[csv['instance'] == 1]
instance2 = csv[csv['instance'] == 2]
instance3 = csv[csv['instance'] == 3]

fig = plt.figure() #plt.figure(figsize=(15,10))

ax = fig.add_subplot(111)
ax.set_title('Throughput', fontsize=18, fontweight='bold')
ax.set_xlabel('Seconds')
ax.set_ylabel('Processed elements per instance')

plt.plot(range(0, len(instance1)), instance1['throughput'], 'r-', label='Instance 1')
plt.plot(range(0, len(instance2)), instance2['throughput'], 'g-', label='Instance 2')
plt.plot(range(0, len(instance3)), instance3['throughput'], 'b-', label='Instance 3')
plt.legend(loc='right')

plt.show()
```
